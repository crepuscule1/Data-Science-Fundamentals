```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.distplot(titanic['fare'], bins=20, kde=False)
plt.ylabel('Frequency')
plt.xlim(0,300)
plt.title('Fare Frequency')
plt.show()
```

```python
sns.boxplot(x='fare', data=titanic)
plt.show()
```

```python
sns.countplot(x='fare', data=titanic)
plt.show()
```

```python
sns.distplot(titanic['fare'], hist=False, rug=True)
plt.show()
```

```python
sns.lmplot(y ='fare', x ='age', data=titanic, hue='survived', legend=True, fit_reg=False) 
plt.show()
```

```python
sns.set(style="whitegrid")
sns.factorplot(x='age', y='fare', hue='sex', data=titanic, col='survived')
plt.show()
```

```python
sns.barplot(x = "age", y = "fare", hue = "survived", data = titanic)
plt.title('Survivors by gender')
plt.show()
```

```python
sns.factorplot(x='sex', y='fare', hue='survived', data=titanic)
plt.show()
```

```python
sns.violinplot(x='sex', y='fare', data=titanic)
plt.show()
```

```python
sns.violinplot(x='sex', y='fare', hue='survived', data=titanic)
plt.show()
```

```python
sns.stripplot(x='sex', y ='fare', data=titanic)
plt.show()
```

```python
sns.boxplot(x='fare', y='sex', data=titanic, whis=np.inf)
sns.stripplot(x='fare', y='sex', data=titanic, jitter=True, color=".3")
plt.show()
```

```python
sns.violinplot(x='sex', y='fare', data=titanic, inner=None, color=".8")
sns.stripplot(x='sex', y='fare', data=titanic, jitter=True)
plt.show()
```
