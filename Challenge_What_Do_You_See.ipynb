```python
import pandas as pd
df = pd.read_csv('Voter_Registration_Data.csv')
print(df)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Party = pd.DataFrame()
Party['Dem'] = [4351, 4293, 4264, 4685, 4267]
Party['Rep'] = [5695, 5587, 5551, 5550, 5538]
Party['Month'] = ['July', 'August', 'September', 'October', 'November']
plt.plot(Party['Month'], Party['Dem'], color ='blue')
plt.plot(Party['Month'], Party['Rep'], color = 'red')
plt.ylabel('Voters Registered')
plt.xlabel('Month')
plt.title('Voter Registration in Curry County Oregon')
plt.legend
plt.show()
```

```python
#The chart above show the number of voters registered by party over a time perios of 5 months in 2018.  The Republican voters have many more registered voters.  There was a spike in Democratic voter registration in October
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1, 2, 1)
plt.plot(Party['Month'],Party['Dem'],color = 'blue')
plt.ylabel('Voters Registered')
plt.xlabel('Month')
plt.title('Democratic Voters')

plt.subplot(1, 2, 2)
plt.plot(Party['Month'], Party['Rep'], color = 'red')
plt.ylabel('Voters Registered')
plt.xlabel('Month')
plt.title('Republican Voters')
plt.show()
```

```python
# The side by side charts show the difference in registration patterns.  There was a decline of Republican voter registration over time between July and November.  The Democrats spiked in October then dropped off in November.
```

```python
plt.hist(Party['Rep'], bins = 5, color = 'red')
plt.hist(Party['Dem'], bins = 5, color = 'blue')
plt.xlabel = ('Votes')
plt.ylabel = ('Months')
plt.title = ('Dem and Rep Voters Registered')
plt.show()
```

```python
#The histogram show the spread of Repubican and Democratic voters.  
```
